```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
#Initialize relative path of csv-file
relpathCar="../MajorAssignment/used-cars-database/autos.csv"
relpathPLZ="../MajorAssignment/PLZ.csv"

#import csv-file into pandas df
carData=pd.read_csv(relpathCar,sep=",",encoding="latin-1", index_col=0)

#show head of carData df
carData.head(5)
```

```python
#Check dtypes
carData.info()
```

```python
#Check NaN values 
#---> Wie wollen wir hier mit fehlenden Werten umgehen? 
carData.isna().sum()
```

```python
#Fill missing values with "-"
carData = carData.fillna("-")
carData.head()
```

```python
#Column dateCreated & lastSeen in Datetime umwandeln 
#dt.normalize() um die Anzeige und nicht dtype selbst zu verändern
carData["dateCreated"] = pd.to_datetime(carData["dateCreated"]).dt.normalize()
carData["lastSeen"] = pd.to_datetime(carData["lastSeen"]).dt.normalize()

#Added Column offerTime to calculate how long the car was online before it was sold 
carData["offerTime"]  = (carData["lastSeen"] - carData["dateCreated"]).dt.days

#Check dtypes
carData.dtypes
```

```python
carData.head(10)
```

```python
#37675 wurden im Monat 0 zugelassen, was machen wir mit denen? 
carData['monthOfRegistration'].value_counts()
```

```python
#Könnten wir ein pie chart drauß machen mit denen > 10.000 (Rest Sonstige)
carData['brand'].value_counts()
```

```python
#12 Gesuche, 371516 Angebote...--> Deshalb drop die 12 Gesuche 
carData = carData[carData.offerType != "Gesuch"]

#drop nrOfPictures, da kein Mehrwert
carData = carData.drop('nrOfPictures', axis=1)
```

```python
#Create df with unique postalCode values from carData
df1 = pd.DataFrame(carData['postalCode'].unique(),  columns=['postalCode']) 
df1
```

```python
#Quelle http://www.fa-technik.adfc.de/code/opengeodb/PLZ.tab
#Alternative wäre geopy gewesen, aber funktioniert bei sovielen Zeilen wegen HTTP429 To many requests nicht

df2 = pd.read_csv(relpathPLZ, sep='\t', encoding="utf-8")
df2.head()
```
